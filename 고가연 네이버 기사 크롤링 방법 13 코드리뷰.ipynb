```python
import sys, os
import requests #(네이버) api 이용할 때 쓰는 패키지
import selenium #크롤링 패키지
from selenium import webdriver
from pandas import DataFrame
from bs4 import BeautifulSoup #크롤링 패키지
import re #정규식 표현 모듈
from datetime import datetime
import pickle, progressbar, json, glob, time #pickle: 객체를 파일로 저장해주는 모듈, progressbar: 진행표시 바, json: 네이버 api, glob: 파일들 리스트 뽑을 때 사용
# https://wikidocs.net/83
from tqdm import tqdm #진행표시바
#

from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.action_chains import ActionChains
#

from wordcloud import WordCloud, STOPWORDS #워클, 불용어 처리
import matplotlib.pyplot as plt
#

from konlpy.tag import Okt #큼직하게 나눔. Mecab은 잘게 나누고 시간도 적게 걸림.
from collections import Counter #글자 수 세주는 모듈
#

from konlpy.utils import pprint
from konlpy.tag import Kkma #대-소 주의, Kkma는 시간이 오래 걸려
#

from konlpy.tag import Twitter #형태소 분석, Okt가 같은 거임.
#

import numpy as np
from PIL import Image #이미지 불러오기
from wordcloud import ImageColorGenerator #이미지 마스크 씌우기
#

import openpyxl #엑셀 파일 불러오기
import pandas as pd
#

import math #수학관련함수
#

import urllib.request #파싱을 위한 모듈
#

import seaborn as sns #시각화 패키지
#


# # 결과 확인을 용이하게 하기 위한 코드
# from IPython.core.interactiveshell import InteractiveShell
# InteractiveShell.ast_node_interactivity = 'all'

## 한글 주석을 위해 추가
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
```

```python
###### 날짜 저장 ##########
date = str(datetime.now())
date = date[:date.rfind(':')].replace(' ', '_')
date = date.replace(':','시') + '분'

sleep_sec = 0.5

####### 언론사별 본문 위치 태그 파싱 함수 ###########
print('본문 크롤링에 필요한 함수를 로딩하고 있습니다...\n' + '-' * 100)
def crawling_main_text(url):

    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36"
    } #headers는 에러 해결을 위해 추가.
    req = requests.get(url, headers = headers) #url에 get요청
    req.encoding = None #깨짐 문제 때문인데, utf-8로 하지 않고 None으로 해줌. 다른 해결방법은 =다음에 'euc-kr'을 쓰는 방법도 있음.
    # http://pythonstudy.xyz/python/article/403-%ED%8C%8C%EC%9D%B4%EC%8D%AC-Web-Scraping
    soup = BeautifulSoup(req.text, 'html.parser') #파씽
    
    # 연합뉴스
    if ('yna' in url) | ('app.yonhapnews' in url): #url에 'yna'또는 'app.yonhapnews'가 들어가면
        main_article = soup.find('div', {'class':'story-news'}) #f12로 해당 페이지에서 기사 본문 위치 찾아서 변수(main_article)에 저장
        if main_article == None: #저 위치에서 본문이 비었다면
            main_article = soup.find('div', {'class' : 'article-txt'}) #다른 위치 변수(main_article)에 저장
            
        text = main_article.text #text추출
    # 나머지 같은 방법으로 진행.
    # MBC 
    elif '//imnews.imbc' in url: 
        text = soup.find('div', {'itemprop' : 'articleBody'}).text
        
    # SBS
    elif 'news.sbs' in url:
        text = soup.find('div', {'itemprop' : 'articleBody'}).text
    
    # KBS
    elif 'news.kbs' in url:
        text = soup.find('div', {'id' : 'cont_newstext'}).text
        
    # JTBC
    elif 'news.jtbc' in url:
        text = soup.find('div', {'class' : 'article_content'}).text
    
    # 그 외 언론사
    else:
        text = "text" #얘네는 아우말이 필요했던 것 (원래 None으로 했었다가 에러떠서 아무말로 바꿔줌.)
    
    return text.replace('\n','').replace('\r','').replace('<br>','').replace('\t','') #text에 많이 나오는 이상한 기호 및 문자들 전부 빈칸으로 대체
    
    
press_list = ['연합뉴스','KBS','MBC','SBS','JTBC'] #없어도 됨.

print('검색할 언론사 : {} | {}개 \n'.format(press_list, len(press_list))) #없어도 됨.


############### 브라우저를 켜고 검색 키워드 입력 ####################
query = input('검색할 키워드  : ')
news_num = int(input('수집 뉴스의 수(숫자만 입력) : '))

print('\n' + '=' * 100 + '\n')

print('브라우저를 실행시킵니다(자동 제어)\n')
browser = webdriver.Chrome('chromedriver') #크롬창열기
# https://beomi.github.io/2017/09/28/HowToMakeWebCrawler-Headless-Chrome/

news_url = 'https://search.naver.com/search.naver?where=news&query={}'.format(query)
browser.get(news_url) #url 오픈
time.sleep(sleep_sec)
```

```python
######### 언론사 선택 및 confirm #####################
print('설정한 언론사를 선택합니다.\n')
search_opt_box = browser.find_element_by_xpath('//*[@id="search_option_button"]')
# 검색 옵션 버튼
search_opt_box.click() #버튼 클릭
time.sleep(0.02)

# 기간 설정 (최근 1주일)
browser.find_element_by_xpath('//*[@id="snb"]/div/ul/li[2]/a').click()
browser.find_element_by_xpath('//*[@id="snb"]/div/ul/li[2]/div/div[1]/ul[1]/li[3]/a').click()

press_box = browser.find_element_by_xpath('//*[@id="snb"]/div/ul/li[5]/a')
# 언론사 버튼 class = "m _tab_option_"인 a 태그
press_box.click()

press_category_box = browser.find_element_by_xpath('//*[@id="order_cat"]')
# class = "press_category"인 div태그의 바로 부모 (id = 'order_cat'인 div태그.)

# browser.find_element_by_xpath('//*[@id="ca_p1"]').click() #첫판 일간지는 선택 안해줘도 됨.
browser.find_element_by_xpath('//*[@id="order_cat"]/div[1]/div/a[2]').click() #방송통신
browser.find_element_by_xpath('//*[@id="ca_1437"]').click() # jtbc
browser.find_element_by_xpath('//*[@id="ca_1056"]').click() # kbs
browser.find_element_by_xpath('//*[@id="ca_1055"]').click() # sbs
browser.find_element_by_xpath('//*[@id="ca_1214"]').click() # mbc
browser.find_element_by_xpath('//*[@id="ca_1001"]').click() # 연합뉴스


# 확인 버튼
press_button  = browser.find_element_by_xpath('//*[@id="snb"]/div/ul/li[5]/div/span/span[1]/button')
press_button.click()

################ 뉴스 크롤링 ########################

print('크롤링을 시작합니다.')
time.sleep(sleep_sec)
# ####동적 제어로 페이지 넘어가며 크롤링
news_dict = {}
idx = 1
cur_page = 1

pbar = tqdm(total=news_num) # 진행표시 바

while idx <= news_num: # idx가 news_num보다 작은게 참인 동안 (커지면 종료)

    table = browser.find_element_by_xpath('//*[@id="main_pack"]/section[1]/div/div[3]/ul') #class가 list_news인 태그 변수 table에 저장
    table_li_list = table.find_elements_by_xpath('.//li[contains(@id, "sp_nws")]') #elements (복수)니까 다 찾는 것, 변수 table 안에서 li 태그 중 id가 sp_nws를 포함하면 가져옴.
    
    li_a_list = [t_li.find_element_by_xpath('.//div/div/a[@class="news_tit"]') for t_li in table_li_list] #table_in_list에서 class 속성이 news_tit인거 하나씩 가져와.
    # https://www.fun-coding.org/crawl_advance5.html에서 [@class=""] 힌트 얻음.

    for n in li_a_list[:min(len(li_a_list), news_num-idx+1)]:
        n_url = n.get_attribute('href')
        # 딕셔너리 생성, 제목, url, 본문
        news_dict[idx] = {'title' : n.get_attribute('title'), 
                          'url' : n_url,
                          'text' : crawling_main_text(n_url)}
        
        idx += 1
        pbar.update(1) #진행바

    if idx < news_num: #크롤링 해야될 뉴스 더 남아있다면
        cur_page +=1 #페이지 넘김
        browser.find_element_by_xpath('//*[@id="main_pack"]/div[3]/div/a[2]').click() #페이지 넘기는 버튼 클릭
        time.sleep(sleep_sec)
    else: #종료
        pbar.close()
        
        print('\n브라우저를 종료합니다.\n' + '=' * 100)
        time.sleep(0.1)
        browser.close()
        break
    browser.implicitly_wait (5)
```

```python
#### 데이터 전처리하기 ###################################################### 

print('데이터프레임 변환\n')
news_df = DataFrame(news_dict).T #행렬 전치

folder_path = os.getcwd() #현재위치(본인 pc기준 다운로드 폴더)에 저장
xlsx_file_name = '네이버뉴스_본문_{}개_{}_{}.xlsx'.format(news_num, query, date)

news_df.to_excel(xlsx_file_name) #엑셀파일로 변환해서 저장.

print('엑셀 저장 완료 | 경로 : {}\\{}\n'.format(folder_path, xlsx_file_name))

os.startfile(folder_path) #디렉토리 열기 (이 코드가 folder_path에 할당돼 있는 디렉토리를 열어줌.)

print('=' * 100 + '\n결과물의 일부')
news_df
```
